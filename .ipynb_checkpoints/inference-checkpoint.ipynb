```python
import os
import pathlib
import tensorflow as tf
import cv2
import time
import numpy as np
from PIL import Image
%matplotlib inline
import matplotlib.pyplot as plt

from object_detection.utils import label_map_util
from object_detection.utils import visualization_utils as viz_utils
```

```python
# Enable GPU dynamic memory allocation
gpu_devices = tf.config.experimental.list_physical_devices('GPU')
for device in gpu_devices:
    tf.config.experimental.set_memory_growth(device, True)
```

```python
# This function will load the saved model provided saved model directory path and return the loaded model.
def load_model(PATH_SAVED_MODEL_DIR):
    '''
    PATH_SAVED_MODEL_DIR :: Input Directory path of saved models.
    Return : Loaded detection model
    '''
    
    print('Loading model...', end='')
    start_time = time.time()

    # LOAD SAVED MODEL AND BUILD DETECTION FUNCTION
    detect_fn = tf.saved_model.load(PATH_SAVED_MODEL_DIR)

    end_time = time.time()
    elapsed_time = end_time - start_time
    print('Done! Took {} seconds'.format(elapsed_time))
    
    return detect_fn    
```

```python
PATH_SAVED_SSD_MODEL_DIR = '/home/ganesh/Desktop/deepak/gun_od/train/exported-models/saved_model'
detect_fn = load_model(PATH_SAVED_SSD_MODEL_DIR)
```

```python
# This function will use the loaded model and 
def detect_labels(IMAGE_PATH, LABEL_MAP_PATH, Model, min_thresh):
    
    category_index = label_map_util.create_category_index_from_labelmap(LABEL_MAP_PATH,
                                                                    use_display_name=True)
    
    print('Running inference for {}... \n'.format(IMAGE_PATH), end='')
    start_time = time.time()
    
    image = cv2.imread(IMAGE_PATH)
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image_expanded = np.expand_dims(image_rgb, axis=0)
    
    # The input needs to be a tensor, convert it using `tf.convert_to_tensor`.
    input_tensor = tf.convert_to_tensor(image)
    
    # The model expects a batch of images, so add an axis with `tf.newaxis`.
    input_tensor = input_tensor[tf.newaxis, ...]

    # input_tensor = np.expand_dims(image_np, 0)
    detections = Model(input_tensor)

    # All outputs are batches tensors.
    # Convert to numpy arrays, and take index [0] to remove the batch dimension.
    # We're only interested in the first num_detections.
    num_detections = int(detections.pop('num_detections'))
    detections = {key: value[0, :num_detections].numpy()
                   for key, value in detections.items()}
    detections['num_detections'] = num_detections

    # detection_classes should be ints.
    detections['detection_classes'] = detections['detection_classes'].astype(np.int64)

    image_with_detections = image.copy()    
    
    # SET MIN_SCORE_THRESH BASED ON YOU MINIMUM THRESHOLD FOR DETECTIONS
    viz_utils.visualize_boxes_and_labels_on_image_array(
          image_with_detections,
          detections['detection_boxes'],
          detections['detection_classes'],
          detections['detection_scores'],
          category_index,
          use_normalized_coordinates=True,
          max_boxes_to_draw=5,
          min_score_thresh=min_thresh,
          agnostic_mode=False)
    
    end_time = time.time()
    elapsed_time = end_time - start_time
    print('Done! Took {} seconds for Detection!!'.format(elapsed_time))
    
    return image_with_detections
```

```python
IMAGE_PATH = '/home/ganesh/Desktop/deepak/gun_od/train/images/test/3.jpeg'
LABEL_MAP_PATH = '/home/ganesh/Desktop/deepak/gun_od/train/annotations/label_map.pbtxt'

image_with_detection = detect_labels(IMAGE_PATH, LABEL_MAP_PATH, detect_fn, 0.4)
```
