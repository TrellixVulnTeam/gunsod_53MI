```python
import os
import random
import shutil
import cv2
```

```python
# Thi function helps us Converting the text file to XML file to train object detection models.
def text_to_xml(source):
    '''
    This function iterates through all the files present in the source directory and genrate xml for those files.
    source directory should have images and their respective annotations as txt
    
    source : type{string}, path of folder.
    return : none
    
    '''
    for filename in os.listdir(source):
        if filename.endswith(".txt"):
            text_path = os.path.join(source, filename)
            image_path = text_path.replace("txt","jpeg")

            with open(text_path,'r',encoding = 'utf-8') as text:
                annot_data = text.read().split()

                annotation_count = int(annot_data[0])

                image = cv2.imread(image_path)
                height,width,depth = image.shape


                mid = ''
                start = f'''
                    <annotation>
                      <filename>{image_path}</filename>
                      <size>
                          <width>{width}</width>
                          <height>{height}</height>
                          <depth>{depth}</depth>
                      </size>
                      <segmented>0</segmented>'''

                for i in range(annotation_count):
                    xmin = int(annot_data[i*4+1])
                    ymin = int(annot_data[i*4+2])
                    xmax = int(annot_data[i*4+3])
                    ymax = int(annot_data[i*4+4])

                    mid += f'''
                              <object>
                                  <name>Gun</name>
                                  <bndbox>
                                      <xmin>{xmin}</xmin>
                                      <ymin>{ymin}</ymin>
                                      <xmax>{xmax}</xmax>
                                      <ymax>{ymax}</ymax>
                                  </bndbox>
                              </object>'''

                end = '''
                    </annotation>
                '''

                final_xml = start + mid + end
                xml_path = text_path.replace("txt","xml")

                with open(xml_path,'w',encoding = 'utf-8') as f:
                    f.write(final_xml)

            print("Done for: ",filename)
            
    return 
```
